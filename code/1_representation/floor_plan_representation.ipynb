```python
# draw a shape based on its vertices
import matplotlib.pyplot as plt

# list of vertices
vertices_1 = [(1, 1), (1, 2), (2, 2), (2, 1)]

# x and y coordinates of the vertices
x1, y1 = zip(*vertices_1)

# add the first vertex to the end to close the shape
x1 = list(x1) + [x1[0]]
y1 = list(y1) + [y1[0]]

# plot the shape
plt.figure()
plt.plot(x1, y1, marker='')

# set the aspect of the plot to be equal
plt.gca().set_aspect('equal', adjustable='box')

plt.axis('off')
plt.show()
```

```python
# irregular shape
vertices_2 = [(1, 1), (2, 1), (2, 2), (3, 2), (3, 0), (1,0)]

x2, y2 = zip(*vertices_2)
x2 = list(x2) + [x2[0]]
y2 = list(y2) + [y2[0]]

plt.figure()
plt.plot(x2, y2, marker='')
plt.gca().set_aspect('equal', adjustable='box')
plt.axis('off')
plt.show()
```

```python
# put 2 shapes in the same plot
plt.figure()
plt.plot(x1, y1, marker='')
plt.plot(x2, y2, marker='')
plt.gca().set_aspect('equal', adjustable='box')
plt.axis('off')
plt.show()
```

```python
# define a function for drawing a shape
def draw_shape(vertices):
    x, y = zip(*vertices)
    x = list(x) + [x[0]]
    y = list(y) + [y[0]]
    plt.plot(x, y, marker='')

# draw the shapes using the function
plt.figure()
draw_shape(vertices_1)
draw_shape(vertices_2)
plt.gca().set_aspect('equal', adjustable='box')
plt.axis('off')
plt.show()
```

## Calculate size using vertices

```python
# house2
room_1 = [(0, 0), (0, 1), (4, 1), (4, 0)]
room_2 = [(0, 1), (0, 5), (4, 5), (4, 1)]
room_3 = [(0, 5), (0, 6), (1.5, 6), (1.5, 5)]
room_4 = [(0, 6), (0, 8.5), (4, 8.5), (4, 7.5), (2.5, 7.5), (2.5, 5), (1.5, 5), (1.5, 6)]
room_5 = [(2.5, 6.5), (2.5, 7.5), (4, 7.5), (4, 6.5)]
room_6 = [(2.5, 5), (2.5, 6.5), (4, 6.5), (4, 5)]

# calculate the area of the house, contains irregular shapes
def calculate_area(vertices):
    x, y = zip(*vertices)
    x = list(x) + [x[0]]
    y = list(y) + [y[0]]
    area = 0
    for i in range(len(x) - 1):
        area += x[i] * y[i + 1] - x[i + 1] * y[i]
    return abs(area) / 2

area = calculate_area(room_1) + calculate_area(room_2) + calculate_area(room_3) + calculate_area(room_4) + calculate_area(room_5) + calculate_area(room_6)

# 4 * 8.5 = 34
print(area)
```
